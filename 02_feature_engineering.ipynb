```python
# -*- coding: utf-8 -*-
# ============================================================================
# NOTEBOOK 02_FEATURE_ENGINEERING - FEATURE ENGINEERING COMPLET (CORRIG√â)
# ============================================================================

# ============================================================================
# CELLULE 1: IMPORTATION DES BIBLIOTH√àQUES
# ============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from pathlib import Path
import warnings
import json
import joblib
from datetime import datetime
import os
warnings.filterwarnings('ignore')

# Configuration
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12

# Chemins
DATA_PATH = Path('dataset.csv')
PROCESSED_DIR = Path('../data/processed')
MODELS_DIR = Path('../models')
REPORTS_DIR = Path('../reports/figures')

# Cr√©ation des dossiers
PROCESSED_DIR.mkdir(exist_ok=True, parents=True)
MODELS_DIR.mkdir(exist_ok=True, parents=True)
REPORTS_DIR.mkdir(exist_ok=True, parents=True)

print("=" * 80)
print("FEATURE ENGINEERING - PR√âPARATION POUR MACHINE LEARNING")
print("=" * 80)
print(f"D√©but: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("=" * 80)

# ============================================================================
# CELLULE 2: CHARGEMENT DES DONN√âES
# ============================================================================

try:
    df = pd.read_csv(DATA_PATH)
    print(f"\n‚úÖ Chargement r√©ussi: {DATA_PATH}")
    print(f"Shape: {df.shape[0]:,} lignes √ó {df.shape[1]} colonnes")
    print(f"Colonnes: {df.columns.tolist()}")
except FileNotFoundError:
    print(f"\n‚ùå Erreur: Fichier {DATA_PATH} non trouv√©!")
    print("üìÅ Cr√©ation d'un dataset d'exemple...")
    
    # Cr√©ation d'un dataset d'exemple
    np.random.seed(42)
    n = 1000
    
    df = pd.DataFrame({
        'sale_price': np.abs(np.random.normal(500000, 150000, n)),
        'year_built': np.random.randint(1950, 2020, n),
        'year_reno': np.random.randint(2000, 2024, n),
        'imp_val': np.abs(np.random.normal(300000, 100000, n)),
        'land_val': np.abs(np.random.normal(200000, 50000, n)),
        'sqft': np.abs(np.random.normal(2000, 500, n)),
        'sqft_lot': np.abs(np.random.normal(5000, 1500, n)),
        'bedrooms': np.random.randint(2, 6, n),
        'bathrooms': np.random.randint(1, 4, n),
        'city': np.random.choice(['Seattle', 'Bellevue', 'Redmond', 'Kirkland'], n),
        'zoning': np.random.choice(['Residential', 'Commercial', 'Industrial', 'Mixed'], n),
        'submarket': np.random.choice(['North', 'South', 'East', 'West'], n),
        'condition': np.random.choice(['Poor', 'Fair', 'Average', 'Good', 'Excellent'], n),
        'grade': np.random.choice(['Low', 'Medium', 'High', 'Premium'], n)
    })
    
    df['year_reno'] = df.apply(lambda row: max(row['year_built'], row['year_reno']), axis=1)
    df.to_csv(DATA_PATH, index=False)
    print(f"‚úÖ Dataset d'exemple cr√©√©: {DATA_PATH}")
    print(f"Shape: {df.shape[0]:,} lignes √ó {df.shape[1]} colonnes")

# ============================================================================
# CELLULE 3: APER√áU DES DONN√âES
# ============================================================================

print("\n" + "=" * 80)
print("APER√áU DES DONN√âES BRUTES")
print("=" * 80)

print("\nüìã 5 premi√®res lignes:")
display(df.head())

print("\nüìä Types de donn√©es:")
print(df.dtypes.value_counts())

print("\nüîç R√©sum√© statistique:")
display(df.describe())

print("\n‚ö†Ô∏è Valeurs manquantes:")
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'Count': missing,
    'Percentage': missing_pct
}).sort_values('Percentage', ascending=False)
display(missing_df[missing_df['Count'] > 0])

# ============================================================================
# CELLULE 4: CR√âATION DES FEATURES D√âRIV√âES
# ============================================================================

print("\n" + "=" * 80)
print("CR√âATION DES FEATURES D√âRIV√âES")
print("=" * 80)

# Ann√©e de vente (ajustez selon vos donn√©es)
CURRENT_YEAR = 2024

# Copie pour feature engineering
df_fe = df.copy()

# 1. Property Age
if 'year_built' in df_fe.columns:
    df_fe['property_age'] = CURRENT_YEAR - df_fe['year_built']
    df_fe['property_age'] = df_fe['property_age'].clip(lower=0)
    print("‚úÖ property_age = ann√©e_vente - year_built")

# 2. Since Reno
if 'year_reno' in df_fe.columns:
    df_fe['since_reno'] = CURRENT_YEAR - df_fe['year_reno']
    df_fe['since_reno'] = df_fe['since_reno'].clip(lower=0)
    print("‚úÖ since_reno = ann√©e_vente - year_reno")

# 3. Improvement/Land Ratio
if all(col in df_fe.columns for col in ['imp_val', 'land_val']):
    df_fe['imp_land_ratio'] = df_fe['imp_val'] / df_fe['land_val']
    df_fe['imp_land_ratio'] = df_fe['imp_land_ratio'].replace([np.inf, -np.inf], np.nan)
    df_fe['imp_land_ratio'] = df_fe['imp_land_ratio'].fillna(df_fe['imp_land_ratio'].median())
    print("‚úÖ imp_land_ratio = imp_val / land_val")

# 4. Sqft/Lot Ratio
if all(col in df_fe.columns for col in ['sqft', 'sqft_lot']):
    df_fe['sqft_ratio'] = df_fe['sqft'] / df_fe['sqft_lot']
    df_fe['sqft_ratio'] = df_fe['sqft_ratio'].replace([np.inf, -np.inf], np.nan)
    df_fe['sqft_ratio'] = df_fe['sqft_ratio'].fillna(df_fe['sqft_ratio'].median())
    print("‚úÖ sqft_ratio = sqft / sqft_lot")

# 5. Log Transformations
if 'sqft' in df_fe.columns:
    df_fe['log_sqft'] = np.log1p(df_fe['sqft'].clip(lower=0))
    print("‚úÖ log_sqft = log(sqft)")

if 'sale_price' in df_fe.columns:
    df_fe['log_price'] = np.log1p(df_fe['sale_price'].clip(lower=0))
    print("‚úÖ log_price = log(sale_price) (pour entra√Ænement)")

print("\nüìä Aper√ßu des features d√©riv√©es:")
features_derivees = ['property_age', 'since_reno', 'imp_land_ratio', 
                     'sqft_ratio', 'log_sqft', 'log_price']
features_derivees = [f for f in features_derivees if f in df_fe.columns]
display(df_fe[features_derivees].head())

# ============================================================================
# CELLULE 5: VISUALISATION DES FEATURES D√âRIV√âES
# ============================================================================

print("\n" + "=" * 80)
print("VISUALISATION DES FEATURES D√âRIV√âES")
print("=" * 80)

if len(features_derivees) > 0:
    n_features = min(len(features_derivees), 6)
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()
    
    for idx, feature in enumerate(features_derivees[:6]):
        ax = axes[idx]
        df_fe[feature].hist(bins=30, ax=ax, edgecolor='black', color='steelblue', alpha=0.7)
        ax.set_title(f'Distribution de {feature}', fontweight='bold')
        ax.set_xlabel(feature)
        ax.set_ylabel('Fr√©quence')
        ax.axvline(df_fe[feature].mean(), color='red', linestyle='--', 
                   label=f'Moyenne: {df_fe[feature].mean():.2f}')
        ax.axvline(df_fe[feature].median(), color='green', linestyle='--', 
                   label=f'M√©diane: {df_fe[feature].median():.2f}')
        ax.legend(fontsize=8)
    
    plt.tight_layout()
    plt.savefig(REPORTS_DIR / 'features_derivees_distributions.png', dpi=100, bbox_inches='tight')
    plt.show()

# ============================================================================
# CELLULE 6: D√âFINITION DES TYPES DE FEATURES
# ============================================================================

print("\n" + "=" * 80)
print("D√âFINITION DES TYPES DE FEATURES")
print("=" * 80)

# Features num√©riques √† normaliser
numerical_features = [
    'sqft', 'sqft_lot', 'bedrooms', 'bathrooms',
    'property_age', 'since_reno', 'imp_land_ratio', 
    'sqft_ratio', 'log_sqft'
]

# Features cat√©gorielles nominales (One-Hot Encoding)
categorical_nominal = ['city', 'zoning', 'submarket']

# Features cat√©gorielles ordinales
categorical_ordinal = ['condition', 'grade']

# Ordres pour l'encodage ordinal
condition_order = ['Poor', 'Fair', 'Average', 'Good', 'Excellent']
grade_order = ['Low', 'Medium', 'High', 'Premium']

# Filtrer les colonnes qui existent
available_cols = df_fe.columns.tolist()
numerical_features = [col for col in numerical_features if col in available_cols]
categorical_nominal = [col for col in categorical_nominal if col in available_cols]
categorical_ordinal = [col for col in categorical_ordinal if col in available_cols]

print(f"\nüìä Features num√©riques √† normaliser ({len(numerical_features)}):")
for col in numerical_features:
    print(f"   - {col}")

print(f"\nüè∑Ô∏è Features nominales (One-Hot) ({len(categorical_nominal)}):")
for col in categorical_nominal:
    print(f"   - {col}")

print(f"\nüìà Features ordinales ({len(categorical_ordinal)}):")
for col in categorical_ordinal:
    print(f"   - {col}")

# ============================================================================
# CELLULE 7: CONFIGURATION DU PR√âPROCESSEUR (CORRIG√âE)
# ============================================================================

print("\n" + "=" * 80)
print("CONFIGURATION DU PR√âPROCESSEUR")
print("=" * 80)

# Pr√©parer les transformateurs
transformers = []

# Transformateur pour les features num√©riques
if numerical_features:
    transformers.append(('num', StandardScaler(), numerical_features))

# Transformateur pour les features nominales (One-Hot)
if categorical_nominal:
    transformers.append(('onehot', OneHotEncoder(
        drop='first', 
        sparse_output=False,
        handle_unknown='ignore'
    ), categorical_nominal))

# Transformateur pour les features ordinales
if categorical_ordinal:
    # D√©finir les cat√©gories uniquement pour les colonnes qui existent
    ordinal_categories = []
    for col in categorical_ordinal:
        if col == 'condition':
            ordinal_categories.append(condition_order)
        elif col == 'grade':
            ordinal_categories.append(grade_order)
    
    transformers.append(('ordinal', OrdinalEncoder(
        categories=ordinal_categories,
        dtype=np.int64,  # Forcer le type entier
        handle_unknown='use_encoded_value',
        unknown_value=-1
    ), categorical_ordinal))

# Cr√©er le pr√©processeur
preprocessor = ColumnTransformer(
    transformers=transformers,
    remainder='drop'
)

print("‚úÖ Pr√©processeur configur√©:")
if numerical_features:
    print(f"   - StandardScaler: {len(numerical_features)} features num√©riques")
if categorical_nominal:
    print(f"   - OneHotEncoder: {len(categorical_nominal)} features nominales")
if categorical_ordinal:
    print(f"   - OrdinalEncoder: {len(categorical_ordinal)} features ordinales")

# ============================================================================
# CELLULE 8: APPLICATION DES TRANSFORMATIONS
# ============================================================================

print("\n" + "=" * 80)
print("APPLICATION DES TRANSFORMATIONS")
print("=" * 80)

# Pr√©paration des donn√©es
feature_cols = numerical_features + categorical_nominal + categorical_ordinal
X = df_fe[feature_cols].copy()  # Copie pour √©viter les probl√®mes

# S'assurer que les colonnes ordinales sont en format string
for col in categorical_ordinal:
    if col in X.columns:
        X[col] = X[col].astype(str)

if 'log_price' in df_fe.columns:
    y = df_fe['log_price']
    y_original = df_fe['sale_price']
    print("‚úÖ Variable cible: log_price")
else:
    y = None
    print("‚ö†Ô∏è Variable cible 'log_price' non disponible")

# Application des transformations
X_processed = preprocessor.fit_transform(X)

# R√©cup√©ration des noms des colonnes
feature_names = []

if numerical_features:
    feature_names.extend(numerical_features)

if categorical_nominal:
    onehot_features = preprocessor.named_transformers_['onehot'].get_feature_names_out(categorical_nominal)
    feature_names.extend(onehot_features)
    print(f"‚úÖ One-Hot Encoding: {len(onehot_features)} colonnes cr√©√©es")

if categorical_ordinal:
    feature_names.extend(categorical_ordinal)

# Cr√©ation du DataFrame final
df_final = pd.DataFrame(X_processed, columns=feature_names)

if y is not None:
    df_final['log_price'] = y.values
    df_final['sale_price_original'] = y_original.values

print(f"\n‚úÖ Transformations appliqu√©es")
print(f"üìä Shape final: {df_final.shape}")
print(f"\nüîç Aper√ßu du dataset final (5 premi√®res lignes):")
display(df_final.head())

# ============================================================================
# CELLULE 9: V√âRIFICATION DE LA NORMALISATION
# ============================================================================

print("\n" + "=" * 80)
print("V√âRIFICATION DE LA NORMALISATION")
print("=" * 80)

if numerical_features:
    print("\nüìä V√©rification StandardScaler (moyenne ‚âà 0, std ‚âà 1):")
    for col in numerical_features[:5]:  # 5 premi√®res features
        mean_val = df_final[col].mean()
        std_val = df_final[col].std()
        print(f"   {col:20s} - moyenne: {mean_val:8.4f}, √©cart-type: {std_val:8.4f}")

if categorical_nominal:
    print(f"\nüè∑Ô∏è One-Hot Encoding - {len(onehot_features)} colonnes")
    print(f"   Exemple: {list(onehot_features[:3])}")

if categorical_ordinal:
    print(f"\nüìà Ordinal Encoding - valeurs uniques:")
    for col in categorical_ordinal:
        if col in df_final.columns:
            unique_vals = sorted(df_final[col].unique())
            print(f"   {col}: {unique_vals}")

print(f"\n‚ö†Ô∏è Valeurs manquantes: {df_final.isnull().sum().sum()}")

# ============================================================================
# CELLULE 10: ANALYSE DES CORR√âLATIONS
# ============================================================================

print("\n" + "=" * 80)
print("ANALYSE DES CORR√âLATIONS")
print("=" * 80)

if 'log_price' in df_final.columns:
    numeric_cols_final = df_final.select_dtypes(include=[np.number]).columns
    corr_matrix = df_final[numeric_cols_final].corr()
    
    # Corr√©lation avec log_price
    price_corr = corr_matrix['log_price'].sort_values(ascending=False)
    
    print("\nüìà Top 10 corr√©lations avec log_price:")
    for col, corr in price_corr.head(11).items():
        if col != 'log_price':
            print(f"   {col:30s}: {corr:.4f}")
    
    # Heatmap
    plt.figure(figsize=(14, 10))
    top_features = price_corr.head(11).index
    sns.heatmap(df_final[top_features].corr(), annot=True, fmt='.2f', 
                cmap='RdBu_r', center=0, square=True, linewidths=0.5)
    plt.title('Matrice de corr√©lation - Top 10 features avec log_price', 
              fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.savefig(REPORTS_DIR / 'correlation_matrix.png', dpi=100, bbox_inches='tight')
    plt.show()

# ============================================================================
# CELLULE 11: SAUVEGARDE DES DATASETS
# ============================================================================

print("\n" + "=" * 80)
print("SAUVEGARDE DES DATASETS")
print("=" * 80)

# 1. Dataset final avec features normalis√©es
final_path = PROCESSED_DIR / 'dataset_final.csv'
df_final.to_csv(final_path, index=False)
print(f"‚úÖ Dataset final: {final_path}")
print(f"   Shape: {df_final.shape[0]:,} lignes √ó {df_final.shape[1]} colonnes")

# 2. Dataset avec prix originaux
if 'log_price' in df_final.columns:
    df_with_price = df_final.drop('log_price', axis=1).copy()
    df_with_price['sale_price'] = np.expm1(df_final['log_price'])
    price_path = PROCESSED_DIR / 'dataset_with_price.csv'
    df_with_price.to_csv(price_path, index=False)
    print(f"‚úÖ Dataset avec prix: {price_path}")

# 3. Sauvegarde du pr√©processeur
preprocessor_path = MODELS_DIR / 'preprocessor.pkl'
joblib.dump(preprocessor, preprocessor_path)
print(f"‚úÖ Pr√©processeur: {preprocessor_path}")

# 4. Sauvegarde des m√©tadonn√©es
metadata = {
    'date': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
    'current_year': CURRENT_YEAR,
    'numerical_features': numerical_features,
    'categorical_nominal': categorical_nominal,
    'categorical_ordinal': categorical_ordinal,
    'condition_order': condition_order,
    'grade_order': grade_order,
    'feature_names': feature_names,
    'shape_original': list(df.shape),
    'shape_final': list(df_final.shape),
    'n_features': len(feature_names)
}

metadata_path = PROCESSED_DIR / 'metadata.json'
with open(metadata_path, 'w', encoding='utf-8') as f:
    json.dump(metadata, f, indent=2, ensure_ascii=False)
print(f"‚úÖ M√©tadonn√©es: {metadata_path}")

# ============================================================================
# CELLULE 12: TEST RAPIDE AVEC RANDOM FOREST
# ============================================================================

print("\n" + "=" * 80)
print("TEST RAPIDE - RANDOM FOREST REGRESSOR")
print("=" * 80)

if 'log_price' in df_final.columns:
    # Pr√©paration des donn√©es
    feature_cols_final = [col for col in df_final.columns 
                         if col not in ['log_price', 'sale_price_original']]
    X_final = df_final[feature_cols_final]
    y_final = df_final['log_price']
    
    # Split train/test
    X_train, X_test, y_train, y_test = train_test_split(
        X_final, y_final, test_size=0.2, random_state=42
    )
    
    print(f"\nüìä Split des donn√©es:")
    print(f"   Train: {X_train.shape[0]:,} lignes")
    print(f"   Test: {X_test.shape[0]:,} lignes")
    print(f"   Features: {X_train.shape[1]}")
    
    # Entra√Ænement
    print("\nüöÄ Entra√Ænement du mod√®le...")
    model = RandomForestRegressor(
        n_estimators=50,
        max_depth=10,
        min_samples_split=5,
        random_state=42,
        n_jobs=-1
    )
    model.fit(X_train, y_train)
    
    # Pr√©dictions
    y_pred = model.predict(X_test)
    
    # M√©triques
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    print("\nüìà R√©sultats:")
    print(f"   RMSE (log): {rmse:.4f}")
    print(f"   RMSE (prix): ${np.expm1(rmse):,.0f}")
    print(f"   R¬≤: {r2:.4f}")
    
    # Importance des features
    importance_df = pd.DataFrame({
        'feature': feature_cols_final,
        'importance': model.feature_importances_
    }).sort_values('importance', ascending=False).head(10)
    
    print("\nüéØ Top 10 features importantes:")
    display(importance_df)
    
    # Visualisation
    plt.figure(figsize=(10, 6))
    plt.barh(importance_df['feature'], importance_df['importance'], 
             color='steelblue', edgecolor='black')
    plt.xlabel('Importance')
    plt.title('Top 10 Features Importantes - Random Forest', fontweight='bold')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.savefig(REPORTS_DIR / 'feature_importance.png', dpi=100, bbox_inches='tight')
    plt.show()
    
    # Sauvegarde du mod√®le
    model_path = MODELS_DIR / 'random_forest_model.pkl'
    joblib.dump(model, model_path)
    print(f"‚úÖ Mod√®le sauvegard√©: {model_path}")

# ============================================================================
# CELLULE 13: RAPPORT FINAL
# ============================================================================

print("\n" + "=" * 80)
print("RAPPORT FINAL - FEATURE ENGINEERING")
print("=" * 80)

print(f"\nüìÅ DONN√âES SOURCE:")
print(f"   ‚Ä¢ Fichier: {DATA_PATH}")
print(f"   ‚Ä¢ Lignes: {df.shape[0]:,}")
print(f"   ‚Ä¢ Colonnes originales: {df.shape[1]}")

print(f"\nüõ†Ô∏è FEATURES CR√â√âES:")
features_crees = [f for f in features_derivees if f in df_fe.columns]
print(f"   ‚Ä¢ Features d√©riv√©es: {len(features_crees)}")
for f in features_crees:
    print(f"     - {f}")

print(f"\n‚öôÔ∏è TRANSFORMATIONS:")
if numerical_features:
    print(f"   ‚Ä¢ Features normalis√©es: {len(numerical_features)}")
if categorical_nominal:
    print(f"   ‚Ä¢ One-Hot Encoding: {len(categorical_nominal)} variables ‚Üí {len(onehot_features) if categorical_nominal else 0} colonnes")
if categorical_ordinal:
    print(f"   ‚Ä¢ Ordinal Encoding: {len(categorical_ordinal)} variables")

print(f"\nüìä DATASET FINAL:")
print(f"   ‚Ä¢ Lignes: {df_final.shape[0]:,}")
print(f"   ‚Ä¢ Colonnes: {df_final.shape[1]}")
print(f"   ‚Ä¢ Features pour ML: {df_final.shape[1] - (2 if 'log_price' in df_final.columns else 0)}")

print(f"\nüíæ FICHIERS SAUVEGARD√âS:")
print(f"   ‚Ä¢ {PROCESSED_DIR / 'dataset_final.csv'}")
if 'log_price' in df_final.columns:
    print(f"   ‚Ä¢ {PROCESSED_DIR / 'dataset_with_price.csv'}")
print(f"   ‚Ä¢ {MODELS_DIR / 'preprocessor.pkl'}")
print(f"   ‚Ä¢ {PROCESSED_DIR / 'metadata.json'}")
print(f"   ‚Ä¢ {REPORTS_DIR}/*.png")

print(f"\n‚úÖ FEATURE ENGINEERING TERMIN√â AVEC SUCC√àS!")
print(f"   Fin: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("=" * 80)

# ============================================================================
# CELLULE 14: V√âRIFICATION FINALE
# ============================================================================

print("\n" + "=" * 80)
print("V√âRIFICATION FINALE")
print("=" * 80)

print("\nüìã Liste des fichiers g√©n√©r√©s:")
for path in [PROCESSED_DIR, MODELS_DIR, REPORTS_DIR]:
    if path.exists():
        files = list(path.glob('*'))
        print(f"\n{path}:")
        for f in files[:5]:  # 5 premiers fichiers
            size = f.stat().st_size / 1024
            print(f"   - {f.name} ({size:.1f} KB)")
        if len(files) > 5:
            print(f"   ... et {len(files) - 5} autres")

print("\n‚úÖ Pr√™t pour le Machine Learning!")
print("=" * 80)
```
